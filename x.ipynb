```python
from datetime import datetime
import random

def random_int():
    return random.randint(1,1000)


def random_string(length: int):
    return "".join(chr(random.randint(97, 122)) for i in range(length))
        
print(random_int())
print(random_string(5))
```

```python
def random_csv(col: dict , n_rows:int) -> pd.DataFrame:
    
    lst = []
    for i in range(n_rows):
        dct = {}
        for j in range (len(col)):
            dct[list(col)[j]] = random_int() if col[list(col)[j]]=="int" else random_string(random_int())
        lst.append(dct)
    df = pd.DataFrame(lst)
    #df.to_csv('output.csv', index=False)
    return df

random_csv({"name": "string", "id":"int"}, 5)
```
